```python
#guess the number in fixed number of try

import random

target = random.randint(1,10)
print(target)
print('Guess a number between 1 to 10\n')
count = 1

while count <=3:
  guess = int(input('your guess '))

  if guess == target:
    print('congratulation!!! you win')
  elif guess < target:
    print('too small. try again!')
  elif guess > target:
    print('too large. try again')
  count += 1

if count > 3:
  print('sorry you lost')
```

```python
#function to guess the integer within the given range by fixed number of try
import random
import math

def guess_number(a, b):
  target = random.randint(a,b)
  print(target)
  print('guess the number between', a, 'and', b, '\n')
  count = 1
  try_limit = int(math.log(b-a+1,2))      #   ??? why ???

  while count <= try_limit:
    print('you can try', try_limit-count+1,'times :')
    guess = int(input('your guess is '))
    if guess == target:
      print('congratulation!!! you won')
    elif guess < target:
      print('too small')
    elif guess > target:
      print('too large')

    count += 1
    
guess_number(1,20)
```

**Maximum Product**

```python

def max_pro(num):
  num = sorted(num, reverse = False)
  print('the list is ', num)

  for i in range(0, len(num), 1):
    prod_1 = num[0]*num[1]*num[-1]
    prod_2 = num[-1]*num[-2]*num[-3]
  print('\nThe maximum product of three numbers is:',max(prod_1,prod_2))


max_pro([1,2,-3,-4])
num = []
```

**Power of Two**

```python
import math

def isPowerofTwo(num):
  if num <=0: 
    return False
  x = math.log10(num)/math.log10(2)
  
  print('power ', x)
  return num == 2++int(x)

isPowerofTwo(25)
```
